```python
#For SVC we need to create N centroids data
#we will use the income vs age scenario

#k clusters with N people

import numpy as np 

def createNkCluster(N , k ):
    pointsPerCluster = float(N)/k
    #Creating two clusters 
    X = []
    Y = []
    for i in range(k):
        incomeCentroidCentre = np.random.uniform(20000, 200000)
        ageCentroidCenter = np.random.uniform(20,70)
        #print(X, Y)
        for i in range(int(pointsPerCluster)):
            X.append([np.random.normal(incomeCentroidCentre,10000), np.random.normal(ageCentroidCenter, 2.0)])
            Y.append(i)           
                       
    X = np.array(X)
    Y = np.array(Y)
    return X,Y        
            
```

```python
%matplotlib inline

from pylab import *

(X,Y) = createNkCluster(100, 5)

plt.figure(figsize = (8,6))
plt.scatter(X[:,0],X[:,1], c= Y.astype(np.float))
plt.show()
```

```python
from sklearn import svm, datasets


C = 1.0
svc = svm.SVC(kernel='linear', C=C).fit(X, Y)
```

```python

#This visualisaton is not terribly important.. Its  a way to see.. Just look at the predictions
def plotPredictions(clf):
    xx, yy = np.meshgrid(np.arange(0, 250000, 10),
                     np.arange(10, 70, 0.5))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    plt.figure(figsize=(8, 6))
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:,0], X[:,1], c=Y.astype(np.float))
    plt.show()
    
    plotPredictions(svc)
```

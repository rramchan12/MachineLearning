```python
#First lets created a random data set and sine of that 
# We will use Decision Tree Regressor - X is assumed to be float and y is the predicted Feature
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

random_generator = np.random.RandomState(2)
x = np.sort(5*random_generator.rand(80,1),axis= 0)


#print (x)
y = np.sin(x).ravel()
#Add some variation

y[::5] += 3 * (0.5 - random_generator.rand(16))

plt.scatter(x,y)

```

```python
from sklearn import tree

reg_tree_with_depth_2 = tree.DecisionTreeRegressor(max_depth=2)
reg_tree_with_depth_5 = tree.DecisionTreeRegressor(max_depth=5)

reg_tree_with_depth_2.fit(x,y)
reg_tree_with_depth_5.fit(x,y)

#Predict
print(np.arange(0.0, 5.0, 0.01).shape)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis,:]
print(np.arange(0.0, 5.0, 0.01)[:, np.newaxis].shape)
#print(X_test)
y_1 = reg_tree_with_depth_2.predict(X_test)
y_2 = reg_tree_with_depth_5.predict(X_test)
```

```python
plt.figure()
plt.scatter(x,y)
plt.plot(X_test, y_1, color="cornflowerblue", label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

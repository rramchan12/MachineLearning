```python
#Data here is extracted from Grouplens-MovieLens Data Set
#There are two files u.data which is the Data, and u.item which shows the name of the movie

# We will create a dataset and merge them, so that its readable (Hmm, This is not for production)

import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']

ratings_df = pd.read_csv("C:/Users/212412835/Documents/Notebooks/ml-100k/u.data", sep = '\t',names = r_cols, usecols= range(3))

m_cols = ['movie_id', 'title']
movies_df = pd.read_csv('C:/Users/212412835/Documents/Notebooks/ml-100k/u.item', sep='|', names=m_cols, usecols=range(2))

ratings_df = pd.merge(movies_df, ratings_df)
```

```python
ratings_df.head()
```

```python
# now we will rearrange to create a pivot table
# Movies List  vs Users 

movie_ratings_pivot_table = ratings_df.pivot_table(index=['user_id'],columns=['title'],values='rating')
movie_ratings_pivot_table.head()
```

```python
#5 rows × 1664 columns - Interesting

#We can extract Users who rated Star Wars

starWarsRatings = movie_ratings_pivot_table['Star Wars (1977)']
starWarsRatings.head()
```

remember covariance - How a variable varies with other variables - 0 to inifinity
Measures how a variable varies in tandem from their means 
	- Eg : Relation between  Age and scores 

 covariance/Std deviation = Corelation = -1 to 1

We do pairwise correlation of Star Wars' vector of user rating with every other movie!

```python
similarMovies = movie_ratings_pivot_table.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```
